# Imports

```python
import pandas as pd
import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
import numpy as np
from numpy import nan
```

# Read-In-Data

```python
Food_poverty = pd.read_excel("C:/Users/Ivette/OneDrive/New folder/DSO 110 Final Project/Correlation Matrix 2011.xlsx")
```

# Univariate Analysis

## Common Data Types

```python
#///////////////////////////////////////////////////#
# Data Types #
#//////////////////////////////////////////////////#

Food_poverty.dtypes
```

```python
Food_poverty.info()
```

```python
Food_poverty.isna()
```

```python
Food_poverty.isnull()
```

```python
#//////////////////////////////////////////#
# Missing Values #
#/////////////////////////////////////////#

msno.matrix(Food_poverty) 
##This shows how many variables are missing or how many rows are missing.
#The White signifies missing and the gray singnifies that the value is there.
##This is a very important step in any EDA(Exploratory Data Anaylsis) you want to make sure you understand exactly what's 
##missing that way as a Data Scienctist or Data Analysist you can determing what you want to do with this in the future!
```

```python
#New Dataframe with only columns with numbers
Food_poverty1 = Food_poverty[['State', 'Food Insecurity Rate', 'Child Food Insecurity Rate', 'Unemployment Rate','2011 Cost Per Meal']]
Food_poverty1.head()
```

# Correlation

```python
Food_poverty1.corr()
```

# Seaborn Correlation Heatmap

```python
plt.figure(figsize=(8,8))
seaborn.heatmap(Food_poverty1.corr(), annot=True, cmap='coolwarm')
```

# Matplotlib Correlation Heatmap

```python
plt.figure(figsize=(16,8))
sns.heatmap(Food_poverty1.corr(), annot=True, annot_kws={"size": 14})
sns.set_style('white')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
