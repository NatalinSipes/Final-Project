# Imports

```python
import pandas as pd
import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
import numpy as np
from numpy import nan
sns.set(color_codes=True)
```

# Read-In-Data

```python
Insufficiencies = pd.read_excel("C:/Users/Ivette/OneDrive/New folder/DSO 110 Final Project/Correlation Matrix 2013(2).xlsx")
```

# Univariate Analysis

## Common Data Types

```python
#///////////////////////////////////////////////////#
# Data Types #
#//////////////////////////////////////////////////#

Insufficiencies.dtypes
```

```python
Insufficiencies.info()
```

```python
Insufficiencies.isna()
```

```python
Insufficiencies.isnull()
```

```python
#//////////////////////////////////////////#
# Missing Values #
#/////////////////////////////////////////#

msno.matrix(Insufficiencies) 
##This shows how many variables are missing or how many rows are missing.
#The White signifies missing and the gray singnifies that the value is there.
##This is a very important step in any EDA(Exploratory Data Anaylsis) you want to make sure you understand exactly what's 
##missing that way as a Data Scienctist or Data Analysist you can determing what you want to do with this in the future!
```

```python
#New Dataframe with only columns with numbers
Insufficiencies1 = Insufficiencies[['State', '2013 Food Insecurity Rate', '2013 Child Food Insecurity Rate', 'Unemployment Rate','2013 Cost Per Meal']]
Insufficiencies1.head()
```

# Correlation

```python
Insufficiencies1.corr()
```

# Seaborn Correlation Heatmap

```python
plt.figure(figsize=(8,8))
seaborn.heatmap(Insufficiencies1.corr(), annot=True, cmap='coolwarm')
```

# Matplotlib Correlation Heatmap

```python
plt.figure(figsize=(16,8))
sns.heatmap(Insufficiencies1.corr(), annot=True, annot_kws={"size": 14})
sns.set_style('white')
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.show()
```
